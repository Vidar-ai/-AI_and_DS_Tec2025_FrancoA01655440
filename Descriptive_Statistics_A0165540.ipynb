#Preparación del entorno

```python
# librerías
import pandas as pd         # Para manejar datos en tablas
import matplotlib.pyplot as plt  # Para graficar
import numpy as np          # Para operaciones numéricas
```

#Cargar el dataset

```python
# Cargar dataset desde la web
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
wine_data = pd.read_csv(url, delimiter=";")

# Mostrar las primeras filas
wine_data.head()
```

#Medidas de Tendencia Centra
Media (Promedio), La media es el promedio de todos los valores.

```python
# Calcular la media d
medias = wine_data.mean()

# Mostrar resultados
print("Medias de todas las variables:")
print(medias)

# Graficar las medias
plt.figure(figsize=(10,5))
medias.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Valores medios de cada característica del vino")
plt.ylabel("Media")
plt.xlabel("Características")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

#Mediana

La mediana es el valor central de los datos ordenados.

```python
# Calcular medianas
medianas = wine_data.median()

print("Medianas de todas las variables:")
print(medianas)

# Graficar medianas
plt.figure(figsize=(10,5))
medianas.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Medianas de cada característica del vino")
plt.ylabel("Mediana")
plt.xlabel("Características")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

#Moda
La moda es el valor que más se repite.

```python
# Calcular la moda (puede haber varias)
modas = wine_data.mode().iloc[0]

print("Modas de todas las variables:")
print(modas)

# Graficar modas
plt.figure(figsize=(10,5))
modas.plot(kind="bar", color="salmon", edgecolor="black")
plt.title("Modas de cada característica del vino")
plt.ylabel("Moda")
plt.xlabel("Características")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

#Medidas de Dispersión
Varianza

```python
varianzas = wine_data.var()

print("Varianzas de cada variable:")
print(varianzas)

# Gráfico
plt.figure(figsize=(10,5))
varianzas.plot(kind="bar", color="orange", edgecolor="black")
plt.title("Varianza de cada característica del vino")
plt.ylabel("Varianza")
plt.xlabel("Características")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

#Desviación Estándar

```python
desviaciones = wine_data.std()

print("Desviación estándar de cada variable:")
print(desviaciones)

# Gráfico
plt.figure(figsize=(10,5))
desviaciones.plot(kind="bar", color="purple", edgecolor="black")
plt.title("Desviación estándar de cada característica del vino")
plt.ylabel("Desviación estándar")
plt.xlabel("Características")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

#Rango


```python
rango = wine_data.max() - wine_data.min()

print("Rango de cada variable:")
print(rango)

# Gráfico
plt.figure(figsize=(10,5))
rango.plot(kind="bar", color="brown", edgecolor="black")
plt.title("Rango de cada característica del vino")
plt.ylabel("Rango")
plt.xlabel("Características")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

#Visualización de Distribuciones
Histograma de Alcohol


```python
wine_data["alcohol"].hist(bins=20, color="lightcoral", edgecolor="black")
plt.title("Histograma del contenido de alcohol en vinos")
plt.xlabel("Alcohol (%)")
plt.ylabel("Frecuencia")
plt.show()
```

#Boxplot de Alcohol

```python
plt.figure(figsize=(6,6))
wine_data.boxplot(column="alcohol")
plt.title("Boxplot del contenido de alcohol en vinos")
plt.ylabel("Alcohol (%)")
plt.show()
```
